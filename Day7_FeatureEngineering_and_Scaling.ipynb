## Feature Scaling
a data preprocessing technique used in machine learning to normalize the range of independent variables (features)

```python
df = pd.DataFrame({
    'Age':[22,34,42,23],
    'Income':[23333,34222,43343,12221]

})
df
```

```python
#Standardization(standard Scaling)---> z-score

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
scaled_data
df1 = pd.DataFrame(scaled_data,columns=df.columns)
df1
```

```python
max(df1['Age']),min(df1['Age'])
```

```python
#Min-Max Scaling (Normalization)


from sklearn.preprocessing import MinMaxScaler

df = pd.DataFrame({
    'Score' : [45,60,75,90]
})
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df)
df2 = pd.DataFrame(scaled_data,columns=df.columns)
df2



```

```python
#Robust Scaling (X - median / IQR)

a= np.mean([1,2,3,4,5])
b = np.mean([1,2,3,4,1000])
a,b


```

```python
from sklearn.preprocessing import RobustScaler
df = pd.DataFrame({
    'Income':[30000,35000,40000,150000]
    })
scaler = RobustScaler()
scaled_data = scaler.fit_transform(df)
df2 = pd.DataFrame(scaled_data,columns = df.columns)
df2
```

```python
#Max Absolute Scaling(-1 to 1) --> X_scaled = X/max(abs(X))

from sklearn.preprocessing import MaxAbsScaler

df = pd.DataFrame({
    'Value':[-5000,-2000,0,3000,8000]
})
scaler = MaxAbsScaler()
scaled_data = scaler.fit_transform(df)
df5 = pd.DataFrame(scaled_data,columns=df.columns)
df5
```

```python
#Log Transformation 
df = pd.DataFrame({
    'Income':[30000,35000,40000,150000]
})
df['Log_income'] = np.log1p(df['Income'])
df
```

```python
#X_transformed = (X^lambda-1)/lambda

from sklearn.preprocessing import PowerTransformer

df = pd.DataFrame({
    'Usage':[5,6,32,12,55]
})

transformer = PowerTransformer(method='yeo-johnson')
transformed_data = transformer.fit_transform(df)
df2 = pd.DataFrame(transformed_data,columns=df.columns)
df2
```
