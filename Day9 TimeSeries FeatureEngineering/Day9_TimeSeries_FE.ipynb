```python
#EMA --> Exponential Moving Average --> 100,120,90 (50,30,20) Explain?
#EMA = (Current Value * (Smoothing Factor)) + (Previous EMA * (1 - Smoothing Factor))
```

```python
df = pd.read_csv('Day9_data.csv')
df
```

```python
df['date'] = pd.to_datetime(df['date'])
df = df.set_index('date')
```

```python
df
```

```python
plt.figure(figsize=(16,9))
plt.plot(df['sales'])
plt.title("Sales over time")
plt.show()
```

```python
df['Moving_Average3'] = df['sales'].shift(1).rolling(3).mean()
df['Moving_Average7'] = df['sales'].shift(1).rolling(7).mean()
df['Moving_Average30'] = df['sales'].shift(1).rolling(30).mean()
```

```python
plt.figure(figsize=(16,9))
plt.plot(df['sales'])
plt.plot(df['Moving_Average3'],label='MA3')
plt.plot(df['Moving_Average7'],label='MA7')
plt.plot(df['Moving_Average30'],label='MA30')
plt.xlabel("Date")
plt.ylabel("Sales")
plt.legend()
plt.show()
```

```python
df['EMA_7'] = df['sales'].shift(1).ewm(span=7).mean()
df['EMA_30'] = df['sales'].shift(1).ewm(span=30).mean()
df
```

```python
plt.plot(df['sales'],label='sales')
plt.plot(df['Moving_Average7'],label='MA7')
plt.plot(df['EMA_7'],label='EMA7')
plt.legend()
plt.show()
```

```python
#difference between MA and EMA is that EMA gives more weight to recent data points, making it more responsive to recent changes in the data, while MA gives equal weight to all data points in the window,
#making it less responsive to recent changes.
```
