```python
# IQR = Q3 - Q1
# Lower Bound = Q1 - 1.5 * IQR
# Upper Bound = Q3 + 1.5 * IQR
```

```python
Q1 = df['total_bill'].quantile(0.25)
Q3 = df['total_bill'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['total_bill'] < lower_bound) | (df['total_bill'] > upper_bound)]

print("Number of outliers: ", len(outliers))
```

```python
sns.boxplot(x=df['total_bill'])
plt.title("Total Bill Boxplot")
plt.show()
```

```python
### Advanced Analysis
```

```python
# Average Tip Percentage by Day
avg_tip_percentage = df.groupby('day')['tip_percentage'].mean()
avg_tip_percentage.plot(kind='bar')
plt.title("Average Tip Percentage by Day")
plt.xlabel("Day")
plt.ylabel("Average Tip Percentage")
plt.show()
```

```python
# Average Tip Percentage by Time
sns.barplot(x='time', y='tip_percentage', data=df)
plt.title("Average Tip Percentage by Time")
plt.xlabel("Time")
plt.ylabel("Average Tip Percentage")
plt.show()
```

Key Insights from EDA
- Total Bill and Tip Have strong positive relationship.
- Dinner generates higher revenue.
- Weekend traffic is high.
- Most customers visit in small groups.
- Presence of Outliers.
- Tip percentage varies across groups.
