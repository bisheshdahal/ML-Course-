```python
# IQR = Q3 - Q1
# Lower Bound = Q1 - 1.5 * IQR
# Upper Bound = Q3 + 1.5 * IQR
```

```python
Q1 = df['total_bill'].quantile(0.25)
Q3 = df['total_bill'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['total_bill'] < lower_bound) | (df['total_bill'] > upper_bound)]

print("Number of outliers: ", len(outliers))
```

```python
# Average Tip Percentage by Time
sns.barplot(x='time', y='tip_percentage', data=df)
plt.title("Average Tip Percentage by Time")
plt.xlabel("Time")
plt.ylabel("Average Tip Percentage")
plt.show()
```

```python
# Visualization of Missing Values
plt.figure(figsize=(10,6))
sns.heatmap(df.isnull(),cbar=False,cmap='viridis')

plt.title("Missing Values Heatmap")
plt.show()
```

```python
# Percentage of Missing Values
missing_percentage = (df.isnull().sum() / len(df)) * 100
print(missing_percentage)
```

```python
df['mass'].fillna(df['mass'].median(), inplace=True)
df['distance'].fillna(df['distance'].median(), inplace=True)
df['orbital_period'].fillna(df['orbital_period'].median(), inplace=True)
```
